```python
#for Target 1 & Target 0 - PIE
def T10pie(name):
    plt.subplots(1,2,figsize=[25,12])
    plt.subplot(1,2,1)
    plt.title(f'Distribution of {name} for those having Payment Difficulties',fontsize=15)
    TARGET_0[name].value_counts().plot.pie(autopct="%1.0f%%")
    plt.subplot(1,2,2)
    plt.title(f'Distribution of {name} for those having Payment Difficulties',fontsize=15)
    TARGET_1[name].value_counts().plot.pie(autopct="%1.0f%%")
    plt.show()

#for Rejected and Approved - PIE
def RejApp_PIE(name):
    plt.subplots(1,2,figsize=[25,12])
    plt.subplot(1,2,1)
    approved_loans[name].value_counts().plot.pie(autopct='%0.1f%%')
    plt.title(f'Distribution of {name} for Approved Applications',fontsize=15)
    plt.subplot(1,2,2)
    rejected_loans[name].value_counts().plot.pie(autopct='%0.1f%%')
    plt.title(f'Distribution of {name} for Rejected Applications',fontsize=15)
    plt.show()
    

#for Target 1 & Target 0 - BAR    
def T10bar(name):

    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,6))
    
    sns.countplot(x=name, data=TARGET_0,ax=ax1)
    ax1.set_ylabel('Total Counts')
    ax1.set_title(f'Distribution of {name} for those having no Payment Difficulties',fontsize=15)
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90, ha="right")
    
    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax1.patches:
        ax1.annotate('{:.1f}%'.format((p.get_height()/len(TARGET_0))*100), (p.get_x()+0.1, p.get_height()+50))
        
    sns.countplot(x=name, data=TARGET_1,ax=ax2)
    ax2.set_ylabel('Total Counts')
    ax2.set_title(f'Distribution of {name} for those having Payment Difficulties',fontsize=15)    
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=90, ha="right")
    
    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax2.patches:
        ax2.annotate('{:.1f}%'.format((p.get_height()/len(TARGET_1))*100), (p.get_x()+0.1, p.get_height()+50))
    
    plt.show()

#for Rejected and Approved - BAR
def RejApp_bar(name):

    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,6))
    
    sns.countplot(x=name, data=approved_loans,ax=ax1)
    ax1.set_ylabel('Total Counts')
    ax1.set_title(f'Distribution of {name} for Approved Applications',fontsize=15)
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90, ha="right")

    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax1.patches:
        ax1.annotate('{:.1f}%'.format((p.get_height()/len(approved_loans))*100), (p.get_x()+0.1, p.get_height()+50))
        
    sns.countplot(x=name, data=rejected_loans,ax=ax2)
    ax2.set_ylabel('Total Counts')
    ax2.set_title(f'Distribution of {name} for Rejected Applications',fontsize=15)    
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=90, ha="right")
    
    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax2.patches:
        ax2.annotate('{:.1f}%'.format((p.get_height()/len(rejected_loans))*100), (p.get_x()+0.1, p.get_height()+50))
    
    plt.show()
```

```python
#by variant analysis function
def birvarcatvsnumplot(cat, num):

    plt.style.use('ggplot')
    sns.despine
    fig,ax = plt.subplots(1,1,figsize=(10,8))
    
    sns.boxenplot(x=cat,y = num, data=merged_ds)
    ax.set_ylabel(f'{num}')
    ax.set_xlabel(f'{cat}')

    ax.set_title(f'{cat} Vs {num}',fontsize=15)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
     
    plt.show()
```
